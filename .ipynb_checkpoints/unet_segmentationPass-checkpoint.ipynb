```python
"""
Helper function for progress bar
"""
def tqdm_enumerate(iterator):
    i = 0
    for y in tqdm(iterator):
        yield i, y
        i += 1
```

#### Augmenating X-rays

```python
path_to_xray = "data\\CXR_png"
path_to_augmented = "data\\CXR_aug"
```

```python
flip_val = [-1,0,1,-1,0,1,-1,0]

"""
def centerCrop(image, t_w, t_h):
        width, height = image.shape
        x1 = int(round((width - t_w) / 2.))
        y1 = int(round((height - t_h) / 2.))
        return image[x1:x1+t_w, y1:y1+t_h]
"""

"""
We keep track of what transformaations has been aplied to the image so that the same transformations could be applied
to their corresponding mask
"""
def augment(image, fileName, savePath, fil_val, angles=None, mask_aug=False):
    if mask_aug is True:
        angles = angles
    else:
        angles = np.random.randint(0, 360, size=8)
    for index, arg in enumerate(angles):
        name_1 = fileName+"_{}.npy".format(index)
        file_name1 = os.path.join(savePath, name_1)
        name_2 = fileName+"_fp_{}.npy".format(index)
        file_name2 = os.path.join(savePath, name_2)
        image_rotated = rotate(image, angle=arg)
        image_flipped = cv2.flip(image_rotated.copy(), flip_val[index])
        if mask_aug is True:
            coords_rotated = np.where(image_rotated != 0)
            image_rotated[coords_rotated] = 1
            coords_flipped = np.where(image_flipped != 0)
            image_flipped[coords_flipped]  = 1 
            np.save(file_name1, image_rotated)
            np.save(file_name2, image_flipped)
        else:
            reshaped_img_rotated = np.reshape(image_rotated, (1, image_rotated.shape[0], image_rotated.shape[1]))
            reshaped_img_flipped = np.reshape(image_flipped, (1, image_flipped.shape[0], image_flipped.shape[1]))
            np.save(file_name1, reshaped_img_rotated)
            np.save(file_name2, reshaped_img_flipped)
    if mask_aug is True:
        return None
    else:
        return angles

org_images = [os.path.join(path_to_xray, f) for f in listdir(path_to_xray) if isfile(join(path_to_xray, f))]
aug_angle_metrics = []

for index, arg in tqdm_enumerate(org_images):
    fileName = arg.split("\\")[-1].split(".png")[0]+'.npy'
    img = io.imread(arg)
    img = img/255
    # Only passing values between 0 and 1
    if img.min() >= 0 and img.max() <= 1:  
        resized = resize(img, (256, 256))
        # Creates 8 different augmentations for each image
        angles = augment(resized.copy(), arg.split("\\")[-1].split(".png")[0], path_to_augmented, flip_val)
        aug_angle_metrics.extend(angles)
        reshaped = np.reshape(resized, (1, resized.shape[0], resized.shape[1]))
        fullPath = os.path.join(path_to_augmented, fileName)
        np.save(fullPath, reshaped)
print("FILE PROCESSING DONE!")
```

#### Augmenting Masks

```python
path_to_left = "data\\ManualMask\\leftMask"
path_to_right = "data\\ManualMask\\rightMask"
combined_mask = "data\\ManualMask\\combinedMask"
combined_mask_aug = "data\\ManualMask\\combinedMaskAug"

left_files = [os.path.join(path_to_left, f) for f in listdir(path_to_left) if isfile(join(path_to_left, f))]
right_files = [os.path.join(path_to_right, f) for f in listdir(path_to_right) if isfile(join(path_to_right, f))]
```

```python
def combine_mask_save(left_mask, right_mask, file, basePath, aug_angle_metrics):
    finalMask = np.add(left_mask, right_mask)
    coords_combined = np.where(finalMask != 0)
    finalMask[coords_combined] = 1
    augment(finalMask.copy(), file, basePath, flip_val, aug_angle_metrics[0:8], True)
    fileName = file +'.npy'
    finalPath = os.path.join(basePath, fileName)
    np.save(finalPath, finalMask)
    del aug_angle_metrics[:8]
    return aug_angle_metrics
    
def read_resize_combine(left_files, right_files, combined_mask_path, aug_angle_metrics):
    for index, arg in tqdm_enumerate(left_files):
        # Loop control
        if index <= len(left_files):
            maskFound = True
            file = arg.split("\\")[-1].split(".png")[0]
            leftImage = resize(io.imread(arg), (256, 256))
            # Try to retrieve right lung if available
            try:
                rightImage = resize(io.imread(right_files[index]), (256, 256))
            except Exception as e:
                maskFound = False
                print(e)
            if maskFound is False:
                rightImage = np.zeros((256, 256))
            aug_angle_metrics = combine_mask_save(leftImage, rightImage, file, combined_mask_path, aug_angle_metrics)        
            """
            fig, ax = plt.subplots(1,2, figsize=(10,10))
            ax[0].imshow(leftImage)
            ax[1].imshow(rightImage)
            plt.show()
            """

read_resize_combine(left_files, right_files, combined_mask_aug, aug_angle_metrics)
print("ALL FILES DONE!")
```

#### Create Train and Validation Set

```python
master_xray = np.load('data_for_Unet/master_xrays.npy')
master_mask = np.load('data_for_Unet/master_masks.npy')
# Training and Testing split
x_train, x_valid, y_train, y_valid = train_test_split(master_xray, master_mask, test_size=0.4, random_state=42)
```

```python
# Transform to torch tensor
tensor_train_x = torch.Tensor(x_train)
tensor_train_y = torch.Tensor(y_train)
tensor_valid_x = torch.Tensor(x_valid)
tensor_valid_y = torch.Tensor(y_valid)

trainset = data.TensorDataset(tensor_train_x, tensor_train_y)
validset = data.TensorDataset(tensor_valid_x, tensor_valid_y)
# Convert to torch dataloaders
trainloader = data.DataLoader(trainset, batch_size=5, shuffle=True)
validloader = data.DataLoader(validset, batch_size=5, shuffle=True)
```

#### Build Model

```python
train_on_gpu = torch.cuda.is_available()

if not train_on_gpu:
    device = torch.device("cpu")
    print("Running on CPU")
else:
    device = torch.device("cuda:0")
    print("CUDA is available!")
```

```python
class UNet(nn.Module):
    
    def encoder(self, in_channels, out_channels, kernel_size=3):
        block = nn.Sequential(nn.Conv2d(in_channels, out_channels, kernel_size=kernel_size, padding=1),
                              nn.ReLU(inplace=True),
                              nn.Dropout(0.25),
                              nn.BatchNorm2d(out_channels),
                              nn.Conv2d(out_channels, out_channels, kernel_size=kernel_size, padding=1),
                              nn.ReLU(inplace=True),
                              nn.Dropout(0.25),
                              nn.BatchNorm2d(out_channels)
                             )
        return block
    
    
    def decoder(self, in_channels, out_channels, kernel_size=3):
        block = nn.Sequential(nn.Conv2d(in_channels, out_channels, kernel_size=kernel_size, padding=1),
                              nn.ReLU(inplace=True),
                              nn.Dropout(0.25),
                              nn.BatchNorm2d(out_channels),
                              nn.Conv2d(out_channels, out_channels, kernel_size=kernel_size, padding=1),
                              nn.ReLU(inplace=True),
                              nn.Dropout(0.25),
                              nn.BatchNorm2d(out_channels),
                              nn.ConvTranspose2d(out_channels, out_channels//2, kernel_size=kernel_size, stride=2,
                                                padding=1, output_padding=1)
                              )
        return block
       

    def final_block(self, in_channels, mid_channel, out_channels, kernel_size=3):
        block = nn.Sequential(nn.Conv2d(in_channels, mid_channel, kernel_size=kernel_size, padding=1),
                              nn.ReLU(inplace=True),
                              nn.Dropout(0.25),
                              nn.BatchNorm2d(mid_channel),
                              nn.Conv2d(mid_channel, mid_channel, kernel_size=kernel_size, padding=1),
                              nn.ReLU(inplace=True),
                              nn.Dropout(0.25),
                              nn.BatchNorm2d(mid_channel),
                              #1X1 CONV
                              nn.Conv2d(mid_channel, out_channels, kernel_size=1, stride=1, padding=0),
                              nn.ReLU(inplace=True),
                              nn.Dropout(0.25),
                              nn.BatchNorm2d(out_channels)
                             )
        return  block
    
    
    """
    def decoder(self, in_channels, mid_channels, out_channels, kernel_size=3):
        block = nn.Sequential(nn.Conv2d(in_channels, mid_channels, kernel_size=kernel_size),
                              nn.ReLU(),
                              nn.BatchNorm2d(mid_channels),
                              nn.Conv2d(mid_channels, mid_channels, kernel_size=kernel_size),
                              nn.ReLU(),
                              nn.BatchNorm2d(mid_channels),
                              nn.ConvTranspose2d(mid_channels, out_channels, kernel_size=3, stride=2,
                                                padding=1, output_padding=1)
                              )
       
    """
    
    def concat_tensors(self, upsampled, bypass, crop=False):
        if crop:
            if bypass.size()[2] != upsampled.size()[2]:
                pad = (bypass.size()[2] - upsampled.size()[2]) // 2
                bypass_padded = F.pad(bypass, (-pad, -pad, -pad, -pad))
                return torch.cat((upsampled, bypass_padded), 1)
            else:
                return torch.cat((upsampled, bypass), 1)
        
    def __init__(self, in_channels, out_channels):
        super().__init__()
        # Encoding Path
        self.conv_encode1 = self.encoder(in_channels, 32)
        self.conv_maxpool1 = nn.MaxPool2d(2,2)
        self.conv_encode2 = self.encoder(32, 64)
        self.conv_maxpool2 = nn.MaxPool2d(2,2)
        self.conv_encode3 = self.encoder(64, 128)
        self.conv_maxpool3 = nn.MaxPool2d(2,2)
        self.conv_encode4 = self.encoder(128, 256)
        self.conv_maxpool4 = nn.MaxPool2d(2,2)
        
        # Bottleneck
        self.bottleNeck = nn.Sequential(nn.Conv2d(256, 512, kernel_size=3, padding=1),
                                        nn.ReLU(inplace=True),
                                        nn.Dropout(0.25),
                                        nn.BatchNorm2d(512),
                                        nn.Conv2d(512, 512, kernel_size=3, padding=1),
                                        nn.ReLU(inplace=True),
                                        nn.Dropout(0.25),
                                        nn.BatchNorm2d(512),
                                        nn.ConvTranspose2d(512, 256, kernel_size=3, stride=2, padding=1, output_padding=1)
                                       )

        # Decoding Path
        self.conv_decode1 = self.decoder(512, 256)
        self.conv_decode2 = self.decoder(256, 128)
        self.conv_decode3 = self.decoder(128, 64)
        self.finalLayer = self.final_block(64, 32, out_channels)
 
    def forward(self, x):
        # Pass through encoding block
        encode_block1 = self.conv_encode1(x)
        encode_pool1 = self.conv_maxpool1(encode_block1)
        encode_block2 = self.conv_encode2(encode_pool1)
        encode_pool2 = self.conv_maxpool2(encode_block2)
        encode_block3 = self.conv_encode3(encode_pool2)
        encode_pool3 = self.conv_maxpool3(encode_block3)
        encode_block4 = self.conv_encode4(encode_pool3)
        encode_pool4 = self.conv_maxpool4(encode_block4)
        bottleNeck = self.bottleNeck(encode_pool4)
        # Pass through decoding block
        concat_path1 = self.concat_tensors(bottleNeck, encode_block4, True)
        decode_block1 = self.conv_decode1(concat_path1)
        concat_path2 = self.concat_tensors(decode_block1, encode_block3, True)
        decode_block2 = self.conv_decode2(concat_path2)
        concat_path3 = self.concat_tensors(decode_block2, encode_block2, True)
        decode_block3 = self.conv_decode3(concat_path3)
        concat_path4 = self.concat_tensors(decode_block3, encode_block1, True)
        final_block = self.finalLayer(concat_path4)
        return final_block
```

```python
model = UNet(1, 2).to(device)
```

```python
from torchsummary import summary
summary(model, (1, 256, 256))
```

```python
import torch.optim as optim

# Specify loss function
criterion = nn.CrossEntropyLoss()
#criterion = nn.BCEWithLogitsLoss()
#criterion = SoftDiceLoss().to(device)

# Specify optimizer
optimizer = optim.SGD(model.parameters(), lr=0.001)
```

#### Train Model

```python
n_epochs = 200
valid_loss_min = np.Inf
train_losses, valid_losses = [], []

for epoch in range(n_epochs):
    train_loss = 0.0
    valid_loss = 0.0
    
    for data, labels in trainloader:
        data, labels = data.to(device, dtype=torch.float), labels.to(device, dtype=torch.long)
        optimizer.zero_grad()
        output = model(data)
        #output = output.permute(0, 2, 3, 1)
        loss = criterion(output, labels)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()*data.size(0)
        
    with torch.no_grad():
        model.eval()
        for data, labels in validloader:
            data, labels = data.to(device, dtype=torch.float), labels.to(device, dtype=torch.long)
            output = model(data)
            #output = output.permute(0, 2, 3, 1)
            loss = criterion(output, labels)
            valid_loss += loss.item()*data.size(0)
            
    model.train()
    train_loss = train_loss/len(trainloader)
    valid_loss = valid_loss/len(validloader)
    train_losses.append(train_loss)
    print("-"*40)
    print("Epoch:{}\t Training Loss:{:.6f}\t Validation Loss: {:.6f}".format(epoch, train_loss, valid_loss))
    
    if valid_loss <= valid_loss_min:
        print("Validation loss decreased ({:.6f} --> {:.6f}).  Saving model ...".format(valid_loss_min, valid_loss))
        # Save Model
        torch.save(model.state_dict(), 'checkpoints/unet_chestCT_4.pth')
        valid_loss_min = valid_loss
```

#### Plot Loss

```python
plt.plot(train_losses, label='Training loss')
plt.plot(valid_losses, label='Validation loss')
plt.legend(frameon=False)
plt.savefig('U-netLoss.png')
```

```python
with torch.no_grad():
    model.eval()
    output = model(images)
    
# Change shape, such that channels comes to the end
output = output.permute(0,2,3,1)
# Get probabilities from logits
softmax = F.softmax(output, dim=1).cpu()
max_val = torch.argmax(softmax, dim=3).cpu().numpy()
image_array = (max_val * 255)
image_array = image_array.astype(np.uint8)
```
